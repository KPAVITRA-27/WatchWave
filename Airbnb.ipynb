# Checking for missing Values

```python
print(df.isnull().sum())
```

```python
df.info()
```

# handling missing values


```python
#last review column convert object to datetime
df['last review']=pd.to_datetime(df['last review'],errors='coerce')
```

```python
df['last review']
```

```python
df.info()
```

```python
#replace NaT and NaN values
df.fillna({'reviews per month':0,'last review':df['last review'].min()},inplace=True)
```

```python
#main details like name nd host name are having nulls drop them
df.dropna(subset=['NAME','host name'],inplace =True)
```

```python
print(df.isnull().sum())
```

```python
#removing columns which r not required
df=df.drop(columns=["license","house_rules"],errors='ignore')
```

```python
df
```

```python
#remove dollar sign and convert to float
df['price']=df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee']=df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

# Removing duplicates from data


```python
df.drop_duplicates(inplace=True)
```

#for description of data

```python
df.describe()
```

  # Visualization

# What is the distribution of listing prices

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50,kde=True,color ='red')
plt.title("Distribution of listing prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

Histogram is evenly distributed of listing prices across different ranges
KDE line helps us to visualize more clearly about the distribution.


# How are different room types distributed


```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="room type", data=df ,color= "blue")
plt.title("Different Room types distribution")
plt.xlabel("Room Type")
plt.ylabel("Count")
plt.show()
```

Most of the people opt for Entire home/apt
Least number of people book hotel rooms

# How are the listings distributed across different neighbourhood

```python
plt.figure(figsize=(12,8))
sns.countplot(y="neighbourhood group", data=df ,color= "green",order=df['neighbourhood group'].value_counts().index)
plt.title("Listings distributed across various neighbourhood")
plt.ylabel("neighbourhood group")
plt.xlabel("Count")
plt.show()
```

# Relationship between Price and room type

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="room type",y="price" ,hue='room type',data=df,palette='Set1')
plt.title("Relationship between price and room type")
plt.xlabel("room type")
plt.ylabel("price")
plt.show()
```

#How has number of views changed over time

```python
df['last review']=pd.to_datetime(df['last review'])
review_over_time=df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(10,6))
review_over_time.plot(kind='line',color='violet')
plt.title("No of reviews changed over time")
plt.xlabel("Date")
plt.ylabel("No of reviews")
plt.show()
```
